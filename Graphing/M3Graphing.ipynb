```python
import pandas as pd 
import matplotlib.pyplot as plt 
import matplotlib.ticker as mtick 
import numpy as np 
%matplotlib inline
from os import getcwd
# print(getcwd())
```

```python
data = pd.read_csv('ConstructionTimeSeriesDataV2.csv')
data.rename(columns={'Month-Year':'Date','Total Construction':'Total','Private Construction':'Private','Public Construction':'Public'},inplace=True)
data.Date = '1-'+data.Date
data.Date = pd.to_datetime(data.Date,dayfirst=True)
data.drop(['Month'],axis=1,inplace=True)
# data.head()
```

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(data.Date[:67],data.Private[:67],c='gray')
ax.plot(data.Date[67:97],data.Private[67:97],label='Private Spending')
ax.plot(data.Date[98:],data.Private[98:],c='gray')
ax.plot(data.Date[:67],data.Public[:67],c='gray')
ax.plot(data.Date[67:97],data.Public[67:97],label='Public Spending')
ax.plot(data.Date[98:],data.Public[98:],c='gray')
ax.set_title('2002-2014 Housing Construction Spending')
ax.set_xlabel('Year')
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}'))
ax.set_ylabel('Construction Spending (in Millions)')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.axvspan(xmin=pd.to_datetime('2007-08-01',dayfirst=True),xmax=pd.to_datetime('2010-01-01',dayfirst=True),facecolor='k',alpha=0.1)
ax.legend()
ax.annotate('2007: Housing\nMarket Crash',
            xy=(pd.to_datetime('2007-08-01',dayfirst=True),80000), xycoords='data',
            xytext=(50,0), textcoords='offset points',
            size=10,
            arrowprops=dict(arrowstyle="simple",
                            fc="0.3", ec="none",
                            connectionstyle="arc3,rad=0.3"))
ax.annotate('2010: Housing Market\nRecovered',
            xy=(pd.to_datetime('2010-01-01',dayfirst=True),40000), xycoords='data',
            xytext=(50,70), textcoords='offset points',
            size=10,
            arrowprops=dict(arrowstyle="simple",
                            fc="0.3", ec="none",
                            connectionstyle="arc3,rad=0.4"))
```

Since we are graphing construction spending, it's important to note the 2007 housing bubble collapse.  The decline in private construction spending lines up nearly perfectly with what most agree as the month the market collapsed, August 2007.  We also can start to see that the market begins to recover in January 2010.  Meanwhile, public construction during the same time period seems to be unaffected.
